```python
test_URL= ('https://raw.githubusercontent.com/ChristianCouch1/HACKTHON-/main/Pedestrian_Signal_Inventory_and_ADA_Self-Assessments_20240402.csv')
test_df = pd.read_csv(test_URL)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
import numpy as np  # Import numpy for NaN checking
sample_size = 500


# Create a random sample of 100 rows
random_sample = test_df.sample(n=sample_size)
# Initialize the Nominatim geocoder
geolocator = Nominatim(user_agent="Hackathon24(2)")


# Initialize an empty list to store the extracted ZIP codes
zip_codes = []


# Loop over each row in the DataFrame
for index, row in random_sample.iterrows():
   # Check for NaN values in latitude or longitude
   if pd.notna(row['EVNT_LAT']) and pd.notna(row['EVNT_LON']):
       # Extract latitude and longitude from the current row
       latitude = row['EVNT_LAT']
       longitude = row['EVNT_LON']


       # Perform reverse geocoding to get the address information
       location = geolocator.reverse((latitude, longitude), exactly_one=True)


       # Extract the ZIP code from the address information
       if location:
           address = location.address
           zip_code = address.split(',')[-2].strip()  # Assumes ZIP code is the second-to-last element in the address
           zip_codes.append(zip_code)
       else:
           zip_codes.append(None)
   else:
       zip_codes.append(None)


# Add the extracted ZIP codes to a new column in the DataFrame
random_sample['ZIP Code'] = zip_codes


# Display the DataFrame with the added 'ZIP Code' column
```
